```python
#importing libraries needed for this project
import os
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing txt file to notebook
soccer = pd.read_csv("goalsscored.txt")
```

```python
# creating an sql database
soccer.to_sql('soccer', sqlite3.connect('goalsscored.db'), if_exists = "replace")
```

```python
#connecting to SQL database
con = sqlite3.connect('goalsscored.db')
#calling database
soccer
```

# First analysis,Looking for the country that scored more than five goals in the  women soccer world cup 2019.

```python
# Creating a first variable as a goal_count to know the number of goals scored by each country.
goal_count = pd.read_sql_query("SELECT Country, Sum(Goals) AS TOTAL_GOALS FROM soccer GROUP BY country HAVING TOTAL_GOALS >= 5 ORDER BY TOTAL_GOALS DESC", con);
                                      
```

```python
# Calling to a new table.
goal_count
```

# The tables shows that the USA is in top of lists followed by England and other country.

# Let graph the above table goal_count using Bar graph.


```python
# lets use a bar graph to plot the above table. I used bar graph because with this we can do comparision between the country in terms of goals they scored.
my_plot = goal_count.plot(kind='bar', color='b', x='COUNTRY', y='TOTAL_GOALS', legend=None,title="Country Scoring At Least 5 Goals In Tournament")
my_plot.set_xlabel("Country")
my_plot.set_ylabel("Total Number Of Goals")
```

# Second analysis, Finding out the number of goals scored by players in terms of their position, For example how many goal scored by forward, midfielder and defender?

```python
position_goal = pd.read_sql_query("SELECT Position, SUM(GOALS) AS POSITION_SCORED FROM soccer GROUP BY position ORDER BY SUM(GOALS) DESC", con);
```

```python
#calling to a another table
position_goal
```

# As we see that forward player scored more goals than midfielder and defender .
Lets graph the table now 

```python
# For displaying this results,I am using the Pie chart beacuse it display all the needed information and also gives pecentages so that we can clearly identified which postion scored most of goal and by how much percent.
plt.figure(figsize = (8, 8))
x = position_goal.POSITION_SCORED
plt.pie(x, autopct='%1.1f%%')
plt.legend(labels = position_goal.POSITION, loc='upper right')
plt.show()
```

```python
# Now with this Piechart,It makes us easier to see the percentage of goals scored by different position players.
```

# My third analysis is to see which country have atleast 4 players who scored a goals for thier respective country.

```python
#Creating a table "country_1" to see the country having atleast 4 players who scored a goals.
country_1 = pd.read_sql_query("SELECT country, COUNT(PLAYER) AS TOTAL_PLAYER FROM soccer GROUP BY country HAVING Total_PLAYER >= 4 ORDER BY COUNT(PLAYER) DESC", con);
```

```python
country_1
```

# Let's plot the results.

```python
# lets use a bar graph to plot the above table. I used bar graph again because with this we can easily do the comparision between the country interms of their number of players who scored a goals.
my_plot = country_1.plot(kind='bar', color='b', x='COUNTRY', y='TOTAL_PLAYER', legend=None,title="Country Having At Least 4 Player Who Scored Goals")
my_plot.set_xlabel("COUNTRY")
my_plot.set_ylabel("NUMBER OF PLAYER WHO SCORED GOALS")
```
